```python
def get_info2(col):
    
    l = list()
    count = 0
    
    for i in ast.literal_eval(col):
        if count < 3:
            l.append(i['name'])
            count += 1
        else:
            break
        
    return l
```

#### Finding the Root Word

#### Stemming

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stemm(col):
    
    l = list()
    
    for i in col.split():
        l.append(ps.stem(i))
        
    return ' '.join(l)
```

```python
df['tags'] = df['tags'].apply(stemm)
```

```python
df
```

#### Finding Similarity

- Cosine Similarity between movies.


- Rows will be like this :-

    - 0th_index_movie [ itself, 1st_index_movie, 2nd_movie_index, ....]

    - 1th_index_movie [ 0st_index_movie, itself, 2nd_movie_index, ....]


- All diagonal elements will be 1.

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
display(similarity)
display(similarity[0])
display(similarity.shape)
```

```python
movie_index = df[df['english_title'] == 'Avatar'].index
display(1, movie_index, 2, df.iloc[movie_index])

distances = similarity[movie_index]
# 2D to 1D
display(3, distances, 4, distances[0])

display(5, list(enumerate(distances)), 6, list(enumerate(distances[0]))[:5])

temp = sorted(list(enumerate(distances[0])),
              reverse = True,
              key = lambda x : x[1]
             )[1:6]

display(7, temp)

row_index_movie = [i[0] for i in temp]
display(8, row_index_movie)

recommended_movie_id = df.loc[row_index_movie, 'movie_id'].to_list()
display(9, recommended_movie_id)

recommended_movie = df.loc[row_index_movie, 'english_title'].to_list()
display(10, recommended_movie)
```

```python
def recommend(movie):
    
    movie_index = df[df['english_title'] == movie].index
    
    distances = similarity[movie_index]
    
    temp = sorted(list(enumerate(distances[0])),
                  reverse = True,
                  key = lambda x : x[1]
                 )[1:6]
    
    row_index_movie = [i[0] for i in temp]
    
    recommended_movie_id = df.loc[row_index_movie, 'movie_id'].to_list()
    
    recommended_movie = df.loc[row_index_movie, 'english_title'].to_list()
    
    
    return recommended_movie_id, recommended_movie
```

```python
display(recommend('Avatar'))
```

```python
display(recommend('Frozen'))
```
